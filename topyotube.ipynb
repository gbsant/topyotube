```python
# importando as bibliotecas
import numpy as np
import pandas as pd
```

```python
# lendo o arquivo topyoutube.csv
df = pd.read_csv('topyoutube.csv')
df.head()
```

```python
# tranformar as colunas sem informacao adicionando o texto “Nao informado”
df = df.fillna("Nao informado")
df
```

```python
# Formatando coluna “AVG” para 2 casas decimais depois da virgula
df['Avg'] = df['Avg'].apply(lambda x: "{:.2f}".format(x))
df
```

```python
# top 10 usuarios
df.head(10)
```

```python
# top 100 artistas por nome decrescente
top100_artists_dec = df.sort_values(by = "Artist" , ascending = False).head(100)
top100_artists_dec
```

```python
# Convertendo o dataframe em um arquivo csv
df.to_csv("topyoutube_transformed.csv", encoding = 'utf-8')
```
